```python
df.columns = df.columns.str.strip()
```

```python
# fill missing values in 'Income' column with median
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
df['Age'] = 2026 - df['Year_Birth']
```

```python
df.duplicated().sum()
```

```python
# check and delete outliers
def check_and_delete_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"total outlier in {column}: {len(outliers)}")
    
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

df = check_and_delete_outliers(df, 'Income')
```

```python
df = check_and_delete_outliers(df, 'Age')
```

```python
# delete absurd values in marital status
df = df[~df['Marital_Status'].isin(['Absurd'])]
```

```python
# stamdardize YOLO and Alone Values in Marital_Status
df['Marital_Status'] = df['Marital_Status'].replace({'YOLO': 'Single', 'Alone': 'Single'})
```

```python
df['Education'] = df['Education'].replace({'2n Cycle': 'Master', 'Graduation': 'Bachelor', 'PhD': 'Doctorate'})
```

### Marketing Analysis

```python
# campaign acceptance rate
acceptance = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] + df['Response']
acceptance_rate = acceptance.sum() / len(df)
acceptance_rate
```

```python
df = df.rename(columns={
    'AcceptedCmp1':'Campaign 1',
    'AcceptedCmp2':'Campaign 2',
    'AcceptedCmp3':'Campaign 3',
    'AcceptedCmp4':'Campaign 4',
    'AcceptedCmp5':'Campaign 5',
    'Response':'Campaign 6'
})
```

```python
# visualization of acceptance per campaign
accepted = ['Campaign 1', 'Campaign 2', 'Campaign 3', 'Campaign 4', 'Campaign 5', 'Campaign 6']
campaign = df[accepted].sum().sort_values(ascending=False)
colors = sns.color_palette("Reds", len(campaign))

plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(campaign, labels=campaign.index, autopct='%1.1f%%', startangle=140, colors=colors)
for text in autotexts:
    text.set_color('white')
    text.set_fontweight('bold')
plt.title('Acceptance per Campaign')
plt.show()
```

```python
# Purchase Frequency by Channel
channels = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
purchase_freq = df[channels].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 6))
sns.barplot(x=purchase_freq.values, y=purchase_freq.index)
plt.title('Purchase Frequency by Channel')
plt.xlabel('Channel')
plt.ylabel('Purchases Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# check what are the most correlated features with Total Spending
# exclude this columns because they are either identifiers, time-related, or directly related to spending and purchases
exclude_cols = ['ID', 'Year_Birth', 'Dt_Customer', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'Campaign 1', 'Campaign 2', 'Campaign 3', 'Campaign 4', 'Campaign 5', 'Campaign 6']

def corr_exclude(df, target, exclude_cols):
    cols = [c for c in df.columns if c not in exclude_cols]
    return df[cols].corr(numeric_only=True)[target].sort_values(ascending=False)

corr_exclude(df, 'Total Spending', exclude_cols)
```

```python
# visualize the correlation between income and spending using scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Income', y='Total Spending', data=df, edgecolor=None)
plt.axhline(df['Total Spending'].mean(), color='grey', linestyle='--')

plt.title('Income vs Total Spending')
plt.xlabel('Income')
plt.ylabel('Total Spending')
plt.show()
```
